```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib as mpl
mpl.rc("savefig", dpi=100)
```

```python
df = pd.DataFrame({
    'Features':['statistical','statistical','statistical','Power','Power','Power','Fractal','Fractal','Fractal','Wavelet','Wavelet','Wavelet'],
    'Classifiers': ['XGBoost','SVM','KNN','XGBoost','SVM','KNN','XGBoost','SVM','KNN','XGBoost','SVM','KNN'],
    'Accuracy' : [79.00,72.52,65.06,67.19,60.92,59.16,63.42,62.95,57.71,69.80,66.23,59.18]
})
```

```python
df
```

```python
sns.set(style="darkgrid")

# Set the figure size
plt.figure(figsize=(12, 8))
plt.title("Valence (Low vs High)")

# Reorder this data frame
# df = df.sort_values(['Accuracy'], ascending=False).reset_index(drop=True)

# Make a barplot
plot = sns.barplot(x="Features", y="Accuracy", hue="Classifiers", data=df, ci=None,palette='Blues_d');

for p in plot.patches:
    plot.annotate('{:.2f}%'.format(p.get_height()), (p.get_x()+0.15, p.get_height()+0.5),
        ha='center', va='bottom',
        color= 'black')
        
plot.get_figure().savefig("valence-scenario1.png")
```
